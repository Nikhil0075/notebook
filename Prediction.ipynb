```python
# Group the data by ticker
groups_by_ticker = b.groupby('Index')
# Check the groups in the dataframe
groups_by_ticker.groups.keys()
```

```python
def train_and_forecast(biomass):
    
    model = Prophet(yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,n_changepoints=1,
    changepoint_range=0.9) 
    
    
    model.fit(biomass)
    
    pred = model.predict(f_d)
    
    error = model.predict(pd.DataFrame(biomass['ds'],columns=['ds']))
    
    mae = mean_absolute_error(biomass['y'],error['yhat'])
    pred = pred[['ds','yhat','trend']].copy()
    pred['mae'] = mae
    pred['Index'] = biomass['Index'].iloc[0]
    return pred
```

```python
n = 2418
output = []
for i in range(n):
    output.append(i)
print(output)
```

```python
prediction = pd.DataFrame()
for ticker in output:
    biomass = groups_by_ticker.get_group(ticker)
    forecast = train_and_forecast(biomass)
    prediction = pd.concat([prediction,forecast])
    print(ticker)
```

```python
prediction = prediction.reset_index(drop=True)
```

```python
prediction 
```

```python
p = pd.DataFrame()
d_2018 = prediction[prediction['ds']== '2018-01-01']
d_2019 = prediction[prediction['ds']=='2019-01-01']
p = pd.concat([p,d_2018,d_2019])
p
```

```python
p.to_csv('prediction_all.csv')
d_2018.to_csv('pred_2018.csv')
d_2019.to_csv('pred_2019.csv')
```
